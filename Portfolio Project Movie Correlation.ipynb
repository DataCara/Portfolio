```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\farad\Downloads\movies.csv')
```

```python
pd.options.mode.chained_assignment = None
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, pct_missing))
```

```python
# Data cleaning
df = df.dropna(how='any',axis=0) 
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
#Update Datatype

df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop Duplicates

df.drop_duplicates()
```

```python
#Scatter plot with bodget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title ('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Films')

plt.show
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data =df, scatter_kws={"color":"red"}, line_kws={"color":"green"})
```

```python
df.corr(method= 'pearson')  #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title ('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show
```

```python
# Look at Company

df_numerical = df

for col_name in df_numerical.columns:
    if(df_numerical[col_name].dtype == 'object'):
        df_numerical[col_name]=df_numerical[col_name].astype('category')
        df_numerical[col_name] = df_numerical[col_name].cat.codes
        
df_numerical.head
```

```python
correlation_matrix = df_numerical.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title ('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show
```

```python
correlation_mat = df_numerical.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```
